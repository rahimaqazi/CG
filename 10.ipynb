### (A) Develop a simple text screen saver using graphics function

```python
import pygame
import random

# Initialize pygame
pygame.init()

# Set screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Text Screen Saver")

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Create a clock object to control the frame rate
clock = pygame.time.Clock()

# Define a function to generate random position and text
def generate_text():
    x = random.randint(0, WIDTH)
    y = random.randint(0, HEIGHT)
    text = "Hello, World!"
    return x, y, text

# Main loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Fill the screen with black color
    screen.fill(BLACK)

    # Generate random position and text
    x, y, text = generate_text()

    # Create a font object
    font = pygame.font.Font(None, 36)

    # Render the text
    text_surface = font.render(text, True, WHITE)

    # Get the text rectangle
    text_rect = text_surface.get_rect()
    text_rect.center = (x, y)

    # Blit the text onto the screen
    screen.blit(text_surface, text_rect)

    # Update the display
    pygame.display.flip()

    # Control the frame rate
    clock.tick(70)

# Quit pygame
pygame.quit()
```

### (B) Perform smiling face animation using graphic functions

```python
import pygame
import sys

# Initialize pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 600, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Smiling Face Animation")

# Define colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
YELLOW = (255, 255, 0)

# Set up clock
clock = pygame.time.Clock()

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(BLACK)

    # Draw face
    pygame.draw.circle(screen, YELLOW, (WIDTH//2, HEIGHT//2), 100)

    # Draw eyes
    pygame.draw.circle(screen, BLACK, (WIDTH//2 - 40, HEIGHT//2 - 30), 15)
    pygame.draw.circle(screen, BLACK, (WIDTH//2 + 40, HEIGHT//2 - 30), 15)

    # Draw mouth (smile)
    pygame.draw.arc(screen, BLACK, (WIDTH//2 - 40, HEIGHT//2 + 10, 80, 40), 0, 3.14, 2)

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(30)

# Quit pygame
pygame.quit()
sys.exit()
```

### (C) Draw the moving car on a screen

```python
import pygame
import sys

# Initialize pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Moving Car")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (128, 128, 128)
RED = (255, 0, 0)

# Car parameters
car_width = 100
car_height = 50
car_x = (WIDTH - car_width) // 2
car_y = HEIGHT - car_height - 20
car_speed = 5

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the car
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        car_x -= car_speed
    if keys[pygame.K_RIGHT]:
        car_x += car_speed

    # Clear the screen
    screen.fill(WHITE)

    # Draw the road
    pygame.draw.rect(screen, GRAY, (0, HEIGHT//2, WIDTH, HEIGHT//2))

    # Draw the car
    pygame.draw.rect(screen, RED, (car_x, car_y, car_width, car_height))
    pygame.draw.circle(screen, BLACK, (car_x + 20, car_y + car_height), 15)
    pygame.draw.circle(screen, BLACK, (car_x + car_width - 20, car_y + car_height), 15)

    # Update the display
    pygame.display.flip()

# Quit pygame
pygame.quit()
sys.exit()
```

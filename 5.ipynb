5a.Develop the program for mid-point circle drawing algorithm

```python
import turtle

def draw_circle(center, radius):
    turtle.speed(2)
    turtle.penup()
    turtle.goto(center[0], center[1] - radius)
    turtle.pendown()

    x, y = 0, radius
    p = 1 - radius

    plot_points(x, y)

    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

        plot_points(x, y)

    turtle.done()

def plot_points(x, y):
    turtle.penup()
    turtle.goto(x, y)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(y, x)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(-x, y)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(-y, x)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(-x, -y)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(-y, -x)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(x, -y)
    turtle.pendown()
    turtle.dot()

    turtle.penup()
    turtle.goto(y, -x)
    turtle.pendown()
    turtle.dot()

if __name__ == "__main__":
    center = (0, 0)
    radius = 100

    draw_circle(center, radius)
```

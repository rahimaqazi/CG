3. Draw the following basic shapes in the center of the screen :
i. Circle ii. Rectangle iii. Square iv. Concentric Circles v. Ellipse vi. Line

```python
from graphics import *

# Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# Get window width and height
width = win.getWidth()
height = win.getHeight()

# Function to get center coordinates
def get_center():
    center_x = width / 2
    center_y = height / 2
    return center_x, center_y  # Return coordinates as separate values

# i. Circle
center_x, center_y = get_center()  # Unpack coordinates
circle = Circle(Point(center_x, center_y), 50)  # Create Point object
circle.draw(win)

# Keep the window open until closed manually
win.getMouse()
win.close()


# Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# Get window width and height
width = win.getWidth()
height = win.getHeight()

# ii. Rectangle
center = get_center()
rectangle = Rectangle(Point(center[0] - 50, center[1] - 25),  # Adjust dimensions
                       Point(center[0] + 50, center[1] + 25))
rectangle.draw(win)

# Keep the window open until closed manually
win.getMouse()
win.close()

# Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# Get window width and height
width = win.getWidth()
height = win.getHeight()

# iii. Square (special case of rectangle)
center = get_center()
square = Rectangle(Point(center[0] - 25, center[1] - 25),  # Set equal dimensions
                   Point(center[0] + 25, center[1] + 25))
square.draw(win)

# Keep the window open until closed manually
win.getMouse()
win.close()


# Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# Get window width and height
width = win.getWidth()
height = win.getHeight()

# iv. Concentric Circles
center = get_center()
for radius in range(25, 75, 25):  # Adjust starting radius and increments
    c = Circle(center, radius)
    c.draw(win)
    
# Keep the window open until closed manually
win.getMouse()
win.close


# Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# v. Ellipse
center = get_center()
ellipse = Oval(center, 50, 25)  # Adjust radius in both dimensions
ellipse.draw(win)

# Keep the window open until closed manually
win.getMouse()
win.close()


#Create a graphics window
win = GraphWin("Shapes in Center", 400, 400)

# vi. Line
center = get_center()
line = Line(Point(100, 50), Point(150, 100))

line.draw(win)

# Keep the window open until closed manually
win.getMouse()
win.close()
```
